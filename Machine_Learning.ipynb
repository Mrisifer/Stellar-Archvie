```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as mb

x = np.array([2,4,6,8,10,12,14,16])
y = np.array([1,3,5,7,11,13,15,17])
m = 2

z = m*x + y

mb.subplot(1,2,1)
mb.plot(x,z)
mb.subplot(1,2,2)
mb.plot(y,z)
mb.xlabel("X-axis")
mb.ylabel("Y-axis")
mb.title("Supervised Learning")
mb.show()

```

```python
import pandas as pd
np.random.seed(0)
X = 2*np.random.rand(1000,1)
Y = 4+3*X+np.random.randn(1000,1)
data = pd.DataFrame({"X":X.flatten(),"Y":Y.flatten()})

mb.scatter(data["X"],data["Y"],color = 'blue')
mb.xlabel('X')
mb.xlabel('Y')
mb.title("Linear Regression Data")
mb.show()
```
